```python
''' 
`bubble_sort` function takes an array as input and returns the sorted array using the bubble sort algorithm.
Big O notation
Time complexity : O(n^2)
Space complexity: O(1)
'''
def bubble_sort(my_array):
  # Get the length of the array
  n = len(my_array)
  # Loop through the array n-1 times
  for i in range(n-1):
    # Initialize a flag to check if any swaps occurred
    swapped = False
    # Loop through the array from 0 to n-i-1
    for j in range(0, n-i-1):
      # Compare the current element with the next element
      if my_array[j] > my_array[j+1]:
        # Swap the elements if the current element is greater than the next element
        my_array[j], my_array[j+1] = my_array[j+1], my_array[j]
        # Set the flag to True to indicate a swap occurred
        swapped = True
    # Break out of the loop if no swaps occurred in the current iteration
    if not swapped:
      break
  # Return the sorted array
  return my_array

# Test the function with an example array
array = [5, 3, 5, 2, 4, 1]
array = [1, 3, 2, 4, 5]
print(bubble_sort(array))
```
