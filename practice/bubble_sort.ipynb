```python
""" Write the bubble sort algorithm"""
```

```python
def bubble_sort(my_array):
    n = len(my_array)
    for i in range(n-1):
        swapped = False
        for j in range(n-i-1):
            if my_array[j] > my_array[j+1]:
                # swap them
                my_array[j], my_array[j+1] = my_array[j+1], my_array[j]
                swapped = True
        if not swapped:
            break
    return my_array

array = [ 1,0.5,6,34,3,2]
print(bubble_sort(array))
```
