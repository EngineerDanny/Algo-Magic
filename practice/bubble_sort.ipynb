```python
""" Write the bubble sort algorithm"""

def bubble_sort(my_array):
    n = len(my_array)
    for i in range(n-1):
        swapped = False
        for j in range(n-i-1):
            if my_array[j] > my_array[j+1]:
                # swap them
                my_array[j], my_array[j+1] = my_array[j+1], my_array[j]
                swapped = True
        if not swapped:
            break
    return my_array

array = [ 1,0.5,6,34,3,2]
print(bubble_sort(array))
```

```python
"""
How can you modify the bubble sort algorithm to sort the array in descending order instead of ascending order?
Just change the greater than sign to less than to sort in descending order instead of ascending order.
"""

def bubble_sort_descending(my_array):
    n = len(my_array)
    for i in range(n-1):
        swapped = False
        for j in range(n-i-1):
            if my_array[j] < my_array[j+1]:
                my_array[j], my_array[j+1] = my_array[j+1], my_array[j]
                swapped = True
        if not swapped:
            break
    return my_array

print(bubble_sort_descending([1,324,2,567,78,90,0.001, 0.1]))
            
```

```python
""" 
How can you use the bubble sort algorithm to sort an array of strings based on their lengths?
"""

def bubble_sort_str(my_array):
    n = len(my_array)
    for i in range(n-1):
        swapped = False
        for j in range(n-i-1):
            if len(my_array[j]) > len(my_array[j+1]):
                swapped = True
                my_array[j], my_array[j+1] = my_array[j+1], my_array[j]
        if not swapped:
            break
    return my_array

bubble_sort_str(["Daniel Agyapong", "Edward Mensah", "Agama", "Peter"])
```

```python
"""
How can you use the bubble sort algorithm to find the kth smallest or largest element in an array?
"""
def find_smallestk_element(my_array, k):
    n = len(my_array)
    for i in range(n - 1):
        swap = False
        for j in range(n - i -1):
            if my_array[j] > my_array[j+1]:
                my_array[j], my_array[j+1] = my_array[j+1], my_array[j]
                swap = True
        if not swap:
            break
    smallest_k = my_array[k - 1]
    largest_k = my_array[-k]
    return smallest_k, largest_k    

# sorted array: [567, 324, 90, 78, 2, 1, 0.1, 0.001]
print(find_smallestk_element([1, 324, 2, 567, 78, 90, 0.001, 0.1], 2))    
```
